

# ABSTRACTION

Loi b k k s t s n sao cho n di nht v chia ht cho 3.

# DECOMPOSITION

- X l kiu chui.
- Bi ton chia ht cho 3.
- Bi ton sp xp.

# PATTERN RECOGNITION

V input l chui s nn cn slice ra thnh tng ch s v chuyn sang integer. 

Yu cu tip theo output l s LN NHT v CHIA HT cho 3.
- Du hiu chia ht cho 3: tng cc ch s ca s  chia ht cho 3. (tm gi l sum_num)
- Tuy nhin, s c nhng ch s lm cho sum_num khng chia ht cho 3 
=> Phi chn  xa n ch s  sum_num vn chia ht cho 3 v s xut ra l s ln nht c th.
- Ch c 3 trng hp khi ly 1 ch s chia 3:
    + Chia ht.
    + D 1.
    + D 2.
- Vy ch cn xem xt s d r khi sum_num/3. Nu r 

# ALGORITHM DESIGN

# IMPLEMENTATION


```python
num = input().strip()


def solve():
    bucket_0 = [] #divisible
    bucket_1 = [] #remained 1
    bucket_2 = [] #remained 2
    sum_num = 0
    for i in num:
        i = int(i)
        sum_num += i
        if i % 3 == 0:
            bucket_0.append(i)
        elif i % 3 == 1:
            bucket_1.append(i)
        else:
            bucket_2.append(i)

    if sum_num % 3 == 1:
        if bucket_1:
            bucket_1.remove(min(bucket_1))
        else:
            bucket_2.remove(min(bucket_2))
            bucket_2.remove(min(bucket_2))

    elif sum_num % 3 == 2:
        if bucket_2:
            bucket_2.remove(min(bucket_2))
        else:
            bucket_1.remove(min(bucket_1))
            bucket_1.remove(min(bucket_1))
    print(''.join(map(str, sorted(bucket_0+bucket_1+bucket_2, reverse=True))))
solve()
```
